2) Построить график функции:

а) y = x2 + 4x + 3;

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = lambda x: x**2 + 4*x + 3

fig = plt.subplots()

x = np.linspace(-20, 16, 100)

plt.title(r'x**2 + 4*x + 3')
plt.plot(x, y(x))
plt.show()
```

б) y = −2 sin 3x;

```python
y = lambda x: -2*(np.sin(3*x))

fig = plt.subplots()

x = np.linspace(-5, 6, 100)

plt.title(r'y = −2 sin 3x')

plt.plot(x, y(x))

plt.show()
```

в) y = |{x} - 1/2|

```python
import math
from math import fabs 

list_x = []
list_y = []

for el in np.arange(1, 10, 0.1):
    x = math.modf(el)
    x_1 = x[0]
    x_2 = x[1]
    x_new = el - x_1
    y = fabs(x_new - 0.5)
    
    list_x.append(x_new)
    list_y.append(y)


plt.title(r'y = |{x} - 1/2|')

plt.plot(list_x, list_y)

plt.show()
```
